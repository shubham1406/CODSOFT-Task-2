Under sampling method is use for unbalanced data

Building a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

```python
legit_sample = legit.sample(n=38)
```

Concatinating two Dtaframes

```python
new_dataset =pd.concat([legit_sample, fraud],axis=0)
```

```python
new_dataset.head()
```

```python
#Splitting the data into Features & Targets
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Spliting the data into Traning data & Testing data

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,stratify = y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training (Logist Regression Model)

```python
model =LogisticRegression()
```

```python
#Training the Logistic Regression Model for Training data
model.fit(x_train,y_train)
```

Model Evaluation

```python
#accuracy on traning data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on Training Data :',training_data_accuracy)
```

```python
#accuracy on testing data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on Testing Data :',testing_data_accuracy)
```
